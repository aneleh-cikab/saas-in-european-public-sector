# This notebook merges tender and country data

```python
# import cleaned data
tender_data = pd.read_csv('../data/cleaned_recoded_tender_data.csv')
tender_data.head()
```

```python
country_data = pd.read_csv('../data/country_data_all_new.csv')

country_data.head()
```

```python
# Merge the two dataframes
# Merge on country and year
# From country_data keep column called value, but rename it to the name of the indicator

df = tender_data.merge(country_data, how='left', left_on=['country', 'year'], right_on=['country', 'year'])
df = df.drop(df.columns[0], axis=1)
df.head()
```

# recoding data for multilevel analysis

```python
# if contract_type is serviced contract, then contract_type_id = 1, else 0

df["contract_type"] = df["contract_type"].apply(lambda x: 1 if x == "Services" else 0)
```

```python
# recode country column to integers

df["country_id"] = df["country"].astype('category')
df["country_id"] = df["country_id"].cat.codes

# recode year column to itnergers

df["year_id"] = df["year"].astype('category')
df["year_id"] = df["year_id"].cat.codes

# if saas is yes, then saas_id = 1, else 0

df["saas"] = df["saas"].apply(lambda x: 1 if x == "Yes" else 0)
```
